```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_csv('3_results.csv')
except FileNotFoundError:
    print(":  '3_results.csv'  .   C++ .")
    exit()

plot_configs = [
    ('Time_sec', 'Time (seconds)', 'Dependence of Time on Message Size'),
    ('Bandwidth_MBs', 'Bandwidth (MB/s)', 'Dependence of Bandwidth on Message Size')
]

for col_name, ylabel, title in plot_configs:
    fig, ax = plt.subplots(figsize=(10, 6))
    

    ax.plot(df['Bytes'], df[col_name], marker='o', linestyle='-', linewidth=2, label='MPI Send/Recv')

    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Message Size (Bytes)', fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.set_xscale('log') 
    
    if col_name == 'Time_sec':
        ax.set_yscale('log')

    ax.grid(True, which="both", ls="--", alpha=0.5)
    ax.legend()

    plt.tight_layout()
    plt.show()
```
