```python
import random
import os

# Конфигурации: (Количество пар векторов, Размерность)
CONFIGS = [
    (1000, 100),
    (1000, 1000),
    (2000, 2000)
]

def generate_file(pairs, dim):
    filename = f"vectors_{pairs}_{dim}.txt"
    
    if os.path.exists(filename):
        print(f"File {filename} already exists. Skipping.")
        return

    print(f"Generating {filename} ({pairs} pairs, dim {dim})...")
    
    with open(filename, "w") as f:
        f.write(f"{pairs} {dim}\n")
        
        for _ in range(pairs):
            vec_a = [f"{random.uniform(-100, 100):.4f}" for _ in range(dim)]
            f.write(" ".join(vec_a) + "\n")
            
            vec_b = [f"{random.uniform(-100, 100):.4f}" for _ in range(dim)]
            f.write(" ".join(vec_b) + "\n")
            
    print(f"Saved {filename}")

if __name__ == "__main__":
    for p, d in CONFIGS:
        generate_file(p, d)
    print("All files generated.")
```
